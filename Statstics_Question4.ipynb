What is correlation? Give an example with a dataset & graphical representation on jupyter Notebook**

**Answer**: 
Correlation is a statistical measure that describes the degree to which two variables change together. It indicates the strength and direction of a linear relationship between two variables.

Correlation values range from -1 to 1:

-1 indicates a perfect negative correlation, meaning that as one variable increases, the other decreases linearly.

1 indicates a perfect positive correlation, meaning that as one variable increases, the other also increases linearly.

0 indicates no correlation, meaning that there is no linear relationship between the variables.

Here's an example using a dataset and a graphical representation in a Jupyter Notebook:

In this example, we have a dataset with heights and weights of individuals. We calculate the correlation coefficient between height and weight using the corr() function. Then, we create a scatter plot to visualize the relationship between height and weight.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example dataset
data = {
    'Height': [150, 165, 170, 175, 180, 185],
    'Weight': [60, 65, 70, 75, 80, 85]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
# Calculate the correlation coefficient between Height and Weight
correlation_coefficient = df['Height'].corr(df['Weight'])

print("\n#Pearson Correlation Coefficient (r):", correlation_coefficient)

# Calculate the correlation matrix for all variables
correlation_matrix = df.corr()
print("correlation_matrix:\n", correlation_matrix)
print("\n#Pearson Correlation Coefficient (r) from  correlation_matrix:", correlation_matrix.iloc[0, 1])
```

```python
# Plot a heatmap to visualize the correlation matrix
plt.figure(figsize=(4, 3))  # Set the figure size
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

# Plotting
plt.figure(figsize=(4, 3))
sns.scatterplot(data=df, x='Height', y='Weight')
plt.title('Scatter plot of Height vs. Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.show()
```
